```
import numpy as np
import pandas as pd
cpath = "/content/drive/My Drive/Study/12th Sem/Data Mining/Lab/LabReport2/"
```

```
#load data
data = pd.read_csv(cpath+"iris.csv", header=None)
data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
data.head()
```

```
#separating features and target
x = np.array(data.drop(['class'],1))
y = np.array(data['class'])
```

```
#spliting traing and test set
from sklearn.model_selection import train_test_split
#from sklearn.preprocessing import StandardScaler
#x = StandardScaler().fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```
#applying KNN on orginal data
from sklearn.neighbors import KNeighborsClassifier
clf1 = KNeighborsClassifier()
clf1.fit(x_train, y_train)
accuracy1 = clf1.score(x_test, y_test)
print(accuracy1)
#print(clf1.predict([[4.7, 3.2, 1.3, 0.2]]))
```

```
#Applying pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
px = np.array(principalDf[['principal component 1', 'principal component 2']])
```

```
#spliting after pca
px_train, px_test, py_train, py_test = train_test_split(px, y, test_size = 0.3)
```

```
#classification after pca
clf2 = KNeighborsClassifier()
clf2.fit(px_train, py_train)
accuracy2 = clf1.score(px_test, py_test)
print(accuracy2)
```
